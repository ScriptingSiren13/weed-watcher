```python
import ultralytics
from ultralytics import YOLO
```

```python
import os
os.environ["CUDA_LAUNCH_BLOCKING"] = "1"
```

```python
model = YOLO('yolov8l.pt') 
```

```python
model.train(
    data=r"C:\Users\AI Server 2\Desktop\weed-detection-app\dataset\data.yaml",
    epochs=120,                   # ✅ full training now
    imgsz=800,
    batch=4,
    name='weed_detect_yolo_v23_finalFULL',
    patience=15,
    lr0=0.0005,
    cos_lr=True,
    lrf=0.01,

    # Advanced optimization
    box=7.5,
    cls=0.5,
    warmup_epochs=3,
    warmup_bias_lr=0.1,
    weight_decay=0.0005,
    label_smoothing=0.01,

    # Augmentations
    mosaic=1.0,
    mixup=0.2,
    hsv_h=0.015,
    hsv_s=0.7,
    hsv_v=0.4,
    flipud=0.25,
    fliplr=0.5
)
```

```python


model = YOLO("runs/detect/weed_detect_yolo_v23_finalFULL2/weights/best.pt")

results = model.predict(
    source=r"C:\Users\AI Server 2\Desktop\weed-detection-app\dataset\test\images",
    conf=0.5,
    augment=True,  # Enables Test-Time Augmentation
    save=True
)
```
